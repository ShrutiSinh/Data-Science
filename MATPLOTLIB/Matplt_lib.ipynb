```python
plt.figure(figsize=(13,6))
x=np.arange(0.,10,0.05)
a=np.cos(x)
b=np.sin(x)
plt.plot(x,a,'g',linewidth=1,linestyle='solid')
plt.plot(x,b,'y',linewidth=2,linestyle='dashed')
plt.plot(x,a+b,'b',linewidth=3,linestyle='dashdot')
plt.plot(x,a-b,'m',linewidth=4,linestyle='dotted')

plt.show()
```

```python
x=115*np.random.randn(100)
plt.figure(figsize=(10,7))
y=np.arange(len(x))
n,edges,p=plt.hist(x,bins=40,histtype='step')
m=0.5*(edges[1:]+edges[:-1])
m=m.tolist()
l=len(m)
m.insert(0,m[0]-20)
m.append(m[l-1]+30)
n=n.tolist()
n.insert(0,0)
n.append(0)
plt.plot(m,n,"-^")
plt.show()
```

# BOX PLOT

            2. lower quartile

            5. higher quartile
